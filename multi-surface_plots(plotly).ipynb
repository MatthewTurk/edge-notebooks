```python
# 3 surfaces

def make_surfs(name):
    coname=path_co+name+'.co.de20_dil.mom0.fits'
    haname=path_ha+name+'_Halpha.fits'
    vname=path_v+name+'_V.fits'
    
    hdu2 = fits.open(get_pkg_data_filename(coname))[0]
    v=fits.getdata(vname)
    co=fits.getdata(coname)
    co_header = fits.getheader(coname)
    ha=fits.getdata(haname)
    ha_header=fits.getheader(haname)
    
    newha_header = FITS_tools.strip_headers.flatten_header(ha_header)

    matched_co, footprint = reproject_interp(hdu2, newha_header)

    con=matched_co/(np.nanmax(matched_co))
    cof=con+2
    
    # defining my own colorscale
    #col1 is linear blue scale, col3 is log blue scale, col2 is yellow-purple for co
    col1=[[0.0, 'rgb(255,255,255)'], [0.15, 'rgb(240,255,255)'], [0.35, 'rgb(135,206,250)'], [0.5,'rgb(30,144,255)'], [0.7,'rgb(0,0,205)'], [0.8,'rgb(0,0,130)'], [1.0, 'rgb(0,0,0)']]
    col2=[[0.0, 'rgb(255,255,224)'], [0.2, 'rgb(255,255,198)'], [0.35, 'rgb(245,210,150)'], [0.5,'rgb(222,184,135)'],[0.58,'rgb(186,85,211)'], [0.68,'rgb(153,50,204)'], [0.8,'rgb(139,0,139)'], [1.0, 'rgb(0,0,0)']]
    col3=[[0.0, 'rgb(255,255,255)'], [1/10000, 'rgb(240,255,255)'],[1/10,'rgb(135,206,250)'], [1.0, 'rgb(0,0,130)']]
    
    for i in range(len(v)):
        for j in range(len(v[0])):
            if v[i,j]<0:
                v[i,j]=0

    if ha.max()/ha.std()>10:
        has=np.sqrt(ha)
        han=has/(has.max())
        haf=han+4
        color_ha=col3
    else:
        han=ha/(ha.max())
        haf=han+4
        color_ha=col1
        
    if v.max()/v.std()>10:
        vs=np.sqrt(v)
        vn=vs/(vs.max())
        vf=vn
    else:
        vn=v/(v.max())
        vf=vn
        
    han_std=han.std()
    vn_std=np.nanstd(vn)
    
    for i in range(len(haf)):
        for j in range(len(haf[0])):
            if haf[i,j]<4+han_std:
                haf[i,j]=np.nan
            
    for i in range(len(vf)):
        for j in range(len(vf[0])):
            if vf[i,j]<vn_std:
                vf[i,j]=np.nan
                
    axis1 = dict(showbackground=False, 
            gridcolor="rgb(230, 230,230)",      
            zerolinecolor="rgb(230, 230,230)", 
            title='RA'
           )
    axis2 = dict(showbackground=False, 
            gridcolor="rgb(230, 230,230)",      
            zerolinecolor="rgb(230, 230,230)", 
            title='DEC'
           )
    axis3 = dict(showbackground=False, 
            gridcolor="rgb(230, 230,230)",      
            zerolinecolor="rgb(230, 230,230)", 
           )
    layout=dict(title=name,
            scene=dict(xaxis=axis1,
                       yaxis=axis2, 
                       zaxis=axis3),
                       margin=dict(l=280,r=25,b=60,t=60),
                       autosize=False,
                       height=700, width=800)
    
    surf1=dict(z=vf, 
           type='surface', 
           name='V-Band', 
           opacity=0.95,
           colorscale='Earth', 
           showlegend=True, 
           reversescale=True, 
           colorbar=go.ColorBar(x=-0.46, title='V',thickness=27,len=0.9,titlefont={'size':18},tickmode='auto',nticks=6, titleside='top'))
                
    surf2=dict(z=haf, 
           showscale=True, 
           opacity=0.8, showlegend=True, type='surface', 
           cauto=False,
           colorscale=color_ha, 
           hoverinfo='all',
           name='H-alpha', 
           reversescale=False, colorbar=go.ColorBar(x=-0.24,thickness=27,len=0.9, title='Ha',titlefont={'size':18},tickmode='auto',nticks=6, titleside='top'))
                
    surf3=dict(z=cof,
           showscale=True, opacity=0.8, 
           showlegend=True, type='surface',
           name='CO', colorscale=col2, 
           reversescale=False, colorbar=go.ColorBar(x=-0.35,thickness=27,len=0.9, title='CO',titlefont={'size':18},tickmode='auto',nticks=6, titleside='top'))
                
    fig=dict(data=[surf1,surf2,surf3], layout=layout)
    py.plot(fig, filename=name ,auto_open=False)
    #embed=tls.get_embed(url)
    #slist=[name, url, embed]
    #listf.append(slist)
```

```python
# 2 surfaces 

def make_2surfs(name):
    haname=path_ha+name+'_Halpha.fits'
    vname=path_v+name+'_V.fits'
    
    v=fits.getdata(vname)
    
    ha=fits.getdata(haname)
    
    
    # defining my own colorscale
    #col1 is linear blue scale, col3 is log blue scale
    col1=[[0.0, 'rgb(255,255,255)'], [0.15, 'rgb(240,255,255)'], [0.35, 'rgb(135,206,250)'], [0.5,'rgb(30,144,255)'], [0.7,'rgb(0,0,205)'], [0.8,'rgb(0,0,130)'], [1.0, 'rgb(0,0,0)']]
    col3=[[0.0, 'rgb(255,255,255)'], [1/10000, 'rgb(240,255,255)'],[1/10,'rgb(135,206,250)'], [1.0, 'rgb(0,0,130)']]
    
    for i in range(len(v)):
        for j in range(len(v[0])):
            if v[i,j]<0:
                v[i,j]=0

    if ha.max()/ha.std()>10:
        has=np.sqrt(ha)
        han=has/(has.max())
        haf=han+2
        color_ha=col3
    else:
        han=ha/(ha.max())
        haf=han+2
        color_ha=col1
        
    if v.max()/v.std()>10:
        vs=np.sqrt(v)
        vn=vs/(vs.max())
        vf=vn
    else:
        vn=v/(v.max())
        vf=vn
        
    han_std=han.std()
    vn_std=np.nanstd(vn)
    
    for i in range(len(haf)):
        for j in range(len(haf[0])):
            if haf[i,j]<2+han_std:
                haf[i,j]=np.nan
            
    for i in range(len(vf)):
        for j in range(len(vf[0])):
            if vf[i,j]<vn_std:
                vf[i,j]=np.nan
                
    axis1 = dict(showbackground=False, 
            gridcolor="rgb(230, 230,230)",      
            zerolinecolor="rgb(230, 230,230)", 
            title='RA'
           )
    axis2 = dict(showbackground=False, 
            gridcolor="rgb(230, 230,230)",      
            zerolinecolor="rgb(230, 230,230)", 
            title='DEC'
           )
    axis3 = dict(showbackground=False, 
            gridcolor="rgb(230, 230,230)",      
            zerolinecolor="rgb(230, 230,230)", 
           )
    layout=dict(title=name,
            scene=dict(xaxis=axis1,
                       yaxis=axis2, 
                       zaxis=axis3),
                       margin=dict(l=5,r=25,b=60,t=60),
                       autosize=False,
                       height=700, width=800)
    
    surf1=dict(z=vf, 
           type='surface', 
           name='V-Band', 
           opacity=0.95,
           colorscale='Earth', 
           showlegend=True, 
           reversescale=True, 
           colorbar=go.ColorBar(x=-0.6, tickmode='auto',nticks=6))
                
    surf2=dict(z=haf, 
           showscale=True, 
           opacity=0.8, showlegend=True, type='surface', 
           cauto=False,
           colorscale=color_ha, 
           hoverinfo='all',
           name='H-alpha', 
           reversescale=False, colorbar=go.ColorBar(x=-0.45,tickmode='auto',nticks=6))
                
    fig=dict(data=[surf1,surf2], layout=layout)
    url=py.plot(fig, filename=name ,auto_open=False)
    embed=tls.get_embed(url)
    slist=[name, url, embed]
    listf.append(slist)
```
